## 12. 1列目をcol1.txtに，2列目をcol2.txtに保存

> 各行の1列目だけを抜き出したものをcol1.txtに，2列目だけを抜き出したものをcol2.txtとしてファイルに保存せよ．確認にはcutコマンドを用いよ．

```
def problem012() -> None:
    """ 最終行で改行していないとコマンドと一致しないので改行のみ追加
    """
    with open(FILENAME) as f:
        buff = f.readlines()

    for col in range(1, 3):
        with open(f'col{col}.txt', 'w') as f:
            f.write('\n'.join(
                line.split('\t')[col - 1] for line in buff
            ))
            f.write('\n')


problem012()
```

```
!cut -f 1 $FILENAME > 'test_col1.txt'
!cut -f 2 $FILENAME > 'test_col2.txt'
!diff --report-identical-files $FILENAME_COL1 test_col1.txt
!diff --report-identical-files $FILENAME_COL2 test_col2.txt
!echo $FILENAME_COL1
!cat $FILENAME_COL1 | head -n 5
!echo $FILENAME_COL2
!cat $FILENAME_COL2 | head -n 5
```

## 13. col1.txtとcol2.txtをマージ

> 12で作ったcol1.txtとcol2.txtを結合し，元のファイルの1列目と2列目をタブ区切りで並べたテキストファイルを作成せよ．確認にはpasteコマンドを用いよ．

```
def problem013() -> None:
    with open('merge.txt', 'w') as f_w, \
            open(FILENAME_COL1) as f_col1, \
            open(FILENAME_COL2) as f_col2:
        for col1, col2 in zip(f_col1, f_col2):
            f_w.write(f'{col1.rstrip()}\t{col2.rstrip()}\n')


problem013()
```

## 16. ファイルをN分割する

> 自然数Nをコマンドライン引数などの手段で受け取り，入力のファイルを行単位でN分割せよ．同様の処理をsplitコマンドで実現せよ．

```
import math


def problem016() -> int:
    num = int(input('input split number: '))

    with open(FILENAME) as f:
        buff = f.readlines()

    file_len = len(buff)
    file_lines = math.ceil(file_len / num)

    for i in range(num):
        start_pos = i * file_lines
        end_pos = (i + 1) * file_lines
        end_pos = end_pos if end_pos <= file_len else file_len
        
        with open(f'out_{i}.txt', 'w') as f:
            f.write(''.join(buff[start_pos:end_pos]))

    return num


num = problem016()
```

```
import math


def problem016_cmd(basename: str) -> None:
    lines = !wc -l $FILENAME
    lines = int(lines[0].split(' ')[0])
    lines = math.ceil(lines / num)

    !split -l $lines $FILENAME -d split-

    for i in range(num):
        filename = f'{basename}_{i}.txt'
        testname = f'split-{i:02}'
        !diff -s $filename $testname


problem016_cmd('out')
```

### 別解

range 部分で等間隔の for が可能なので、そっちのほうが記述量が少ないです。

```
import math


def problem016_1() -> int:
    num = int(input('input split number: '))

    with open(FILENAME) as f:
        buff = f.readlines()

    file_len = len(buff)
    file_lines = math.ceil(file_len / num)

    for i, start_pos in enumerate(range(0, file_len, file_lines)):
        with open(f'out_1_{i}.txt', 'w') as f:
            f.write(''.join(buff[start_pos:start_pos + file_lines]))

    return num


num = problem016_1()
```

```
problem016_cmd('out_1')
```

## 17. １列目の文字列の異なり

> 1列目の文字列の種類（異なる文字列の集合）を求めよ．確認にはsort, uniqコマンドを用いよ．

```
import functools
from pprint import pprint

def problem017() -> None:
    with open(FILENAME) as f:
        buff = f.readlines()

    str_kind = sorted(set([word.split('\t')[0] for word in buff]))

    with open('uniq.txt','w') as f:
        f.write('\n'.join(str_kind))
        f.write('\n')


problem017()
```

```
!cat $FILENAME | cut --fields 1 | sort | uniq > uniq_test.txt

!diff -s uniq.txt uniq_test.txt
```

## 18. 各行を3コラム目の数値の降順にソート

> 各行を3コラム目の数値の逆順で整列せよ（注意: 各行の内容は変更せずに並び替えよ）．確認にはsortコマンドを用いよ（この問題はコマンドで実行した時の結果と合わなくてもよい）．

この問題ではコマンドの実行結果と順序が一致しないため、単に実行結果を出力するだけにしている。

```
import numpy as np


def problem018() -> None:
    target = 2

    with open(FILENAME) as f:
        buff = f.readlines()

    mat = np.array([word.rstrip().split('\t') for word in buff])
    mat = mat[np.argsort(mat[:, target])[::-1], :]
    pprint(mat)


problem018()
```

```
!cat $FILENAME | sort --key=3,3 --numeric-sort --reverse | head -n 5
```

### 別解

numpy 使わなくても sort の関数指定でできる。

```
from pprint import pprint


def problem018_1() -> None:
    target = 2

    with open(FILENAME) as f:
        buff = f.readlines()

    buff.sort(
        key=lambda line: float(line.split('\t')[target]),
        reverse=True
    )

    pprint(buff)


problem018_1()
```
